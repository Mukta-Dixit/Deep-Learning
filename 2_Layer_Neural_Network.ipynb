### Initialize variables

```python
input_size=x.shape[0]
hiddenL_size1=20
hiddenL_size2=15
output_size=10
```

### Weights and bais values

```python
w1=np.random.randn(hiddenL_size1,input_size)*.01
w1.shape
```

```python
w2=np.random.randn(hiddenL_size2,hiddenL_size1)*.01
w2.shape
```

```python
w3=np.random.randn(output_size,hiddenL_size2)*.01
w3.shape
```

```python
b1=np.zeros((hiddenL_size1,1))
b1.shape
```

```python
b2=np.zeros((hiddenL_size2,1))
b2.shape
```

```python
b3=np.zeros((output_size,1))
b3.shape
```

### Relu function

```python
def relu(r):
    return np.maximum(0,r)
    
```

### Derivative of Relu function

```python
def derivative_Relu(z):
    if z>0:
        return 1
    else:
        return 0
```

### Sigmoid function

```python
def sigmoid(z):
    return (1/ (1 + np.exp(-z)))
```

### Derivative of sigmoid function

```python
def derivative_Sigmoid(z):
    return (z*(1-z))
```

### Forward propogation and Backward propogation

```python
#Forward propogation
for i in range(100):
        z1=np.dot(w1,x)+b1
        A1= relu(z1)

        z2=np.dot(w2,A1)+b2
        A2= relu(z2)

        z3=np.dot(w3,A2)+b3
        y_out= sigmoid(z3)

        #cost function
        m=x.shape[1]
        cost_function=(-1/m)*np.sum( (np.multiply(y.T,np.log(y_out))+ np.multiply((1-y.T),np.log(1-y_out))))
        print(cost_function)

        #Backward propogation

        #3rd layer
        differential_z3= y_out-y.T
        differential_w3= (np.matmul(differential_z3,A2.T)* (1/m))
        differential_b3= np.sum(differential_z3,axis=1,keepdims= True)

        #2nd layer
        differential_second_HiddenL=np.ones((hiddenL_size2,m))
        for i in range(hiddenL_size2):
            for j in range(m):
                 differential_second_HiddenL[i,j]= derivative_Relu(z2[i,j])
        differential_z2= np.dot(w3.T,differential_z3)*differential_second_HiddenL
        differential_w2= (np.matmul(differential_z2,A1.T)* (1/m))
        differential_b2= np.sum(differential_z2,axis=1,keepdims= True)

        #1st layer
        differential_first_HiddenL=np.ones((hiddenL_size1,m))
        for i in range(hiddenL_size1):
            for j in range(m):
                 differential_first_HiddenL[i,j]= derivative_Relu(z1[i,j])
        differential_z1= np.dot(w2.T,differential_z2)*differential_first_HiddenL
        differential_w1=(np.matmul(differential_z1,x.T)* (1/m))
        differential_b1= np.sum(differential_z1,axis=1,keepdims= True)

        #Updating weights and bais
        learning_rate=0.001

        w1=w1-learning_rate*differential_w1
        b1=b1-learning_rate*differential_b1
        w2=w2-learning_rate*differential_w2
        b2=b2-learning_rate*differential_b2
        w3=w3-learning_rate*differential_w3
        b3=b3-learning_rate*differential_b3
```
